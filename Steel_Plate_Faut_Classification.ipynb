# Splitting the dataset into "Features" and "Categories of Fault"

```python
df.info()
```

Clearing spaces from the names of the columns if any

```python
df.columns = df.columns.to_series().apply(lambda x: x.strip())
```

Creating "Faults" from the dataframe

```python
faults =df[["Pastry","Z_Scratch","K_Scatch","Stains","Dirtiness","Bumps","Other_Faults"]]
```

```python
faults.info()
```

Removing the "Faults" from original dataframe and keeping only features

```python
df.drop(df.columns[27:33], axis=1, inplace=True)
```

```python
x = pd.DataFrame(df)
```

```python
x.info()
```

# Converting One-hot Encoding to Single Column Features for Machine Learning Algorithms

```python
y = []
for i in range(faults.shape[0]):
    if faults["Pastry"].values[i] == 1:
        y.append("Pastry")
    elif faults["Z_Scratch"].values[i] == 1:
        y.append("Z_Scratch")
    elif faults["K_Scatch"].values[i] == 1:
        y.append("K_Scatch")
    elif faults["Stains"].values[i] == 1:
        y.append("Stains")
    elif faults["Dirtiness"].values[i] == 1:
        y.append("Dirtiness")
    elif faults["Bumps"].values[i] == 1:
        y.append("Bumps")
    else:
        y.append("Other_Faults")
```

```python
y=np.array(y)
```

```python
y.shape
```

```python
faultstype= pd.DataFrame({'faults':y})
```

```python
faultstype.describe()
```

# Data Standardization

Using variables without standardization in effect gives the variable with the larger range a larger weight in the analysis. Transforming the data to comparable scales can prevent this problem.

```python
from sklearn.preprocessing import StandardScaler
```

```python
X=sc.fit_transform(x)
```
